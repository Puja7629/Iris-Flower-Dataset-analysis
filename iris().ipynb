```python
# loading dataset

iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
# importig libraries

%matplotlib inline
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
features = iris.data[:, [0, 1, 2, 3]]
target = iris.target

features_sum = []
for observations in features:
    features_sum.append(observations[0] + observations[1] + observations[2] + observations[3])
```

```python
# making dataframe of iris dataset

iris_pd = pd.DataFrame({iris.feature_names[0] : iris.data[:,0], iris.feature_names[1] : iris.data[:,1], 
                       iris.feature_names[2] : iris.data[:,2], iris.feature_names[3] : iris.data[:,3],
                        'target' : iris.target, 'species' : 'iris-setosa'})

iris_pd['species'].iloc[50:100] = 'iris-versicolor'
iris_pd['species'].iloc[100:150] = 'iris-virginica'
```

```python
# findout unique classification/type of iris flower.

print(iris_pd['species'].unique())
```

```python
# findout no of rows for each Species.

print(iris_pd.groupby('species').size())
```

```python
# Create 3 DataFrame for each Species

setosa = iris_pd[iris_pd.species == 'iris-setosa']
versicolor = iris_pd[iris_pd.species == 'iris-versicolor']
virginica = iris_pd[iris_pd.species == 'iris-virginica']

print(setosa.describe())
print(versicolor.describe())
print(virginica.describe())
```

```python
#count tells that all the 4 features have 150 rows
#in general ,From Mean we can say that sepal is larger than petal.

print(iris_pd.describe())
```

```python
# Plotting the Scatter Plot of Features_sum and target

plt.figure(figsize = (10, 8))
plt.scatter(features_sum, target, color='red', alpha =1.0)
plt.title('Iris Dataset Features and Target Scatter Plot')
plt.xlabel('Features')
plt.ylabel('Targets')
```

```python
#Plotting Petal Length vs Petal Width & Sepal Length vs Sepal width using Matplotlib.pyplot 

sepal_length = iris_pd['sepal length (cm)']
sepal_width = iris_pd['sepal width (cm)']
petal_length = iris_pd['petal length (cm)']
petal_width = iris_pd['petal width (cm)']

plt.figure(figsize = (18, 8))
plt.subplot(121)
plt.scatter(sepal_length, sepal_width, label = 'Sepal', color = 'r')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Comparison')
plt.legend()
plt.subplot(122)
plt.scatter(petal_length, petal_width, label = 'Petal')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Comparison')
plt.legend()
```
